# Algorithum Grpahs:
This a quick grpahing function for the algorithum homework. This was completed in colab since the environment comes pre installed with basic data analysis tooling (pandas, matplotlib, seaborn ect...)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_style("dark")


df = pd.read_csv("/content/drive/MyDrive/tableConvert.com_gutt5q.csv")
df = df.melt(id_vars=["N"]).rename(columns = {"variable": "Algorithum", "value": "Time (sec)"})



fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

x = "N"
plt.yscale('log')
sns.lineplot(x=df[x], y=df["Time (sec)"], hue =df["Algorithum"], ax = axes[0])
axes[0].set_title("Algorithum: Size vs Time")

sns.lineplot(x=df[x], y=df["Time (sec)"], hue =df["Algorithum"], ax = axes[1])
axes[1].set_title("Algorithum: Size vs log(Time)")
axes[1].set_yscale('log')
plt.show()
```
